```python
def plot_properties(data):
    ave_inflammations = data.mean(axis=0)
    max_inflammations = data.max(axis=0) 
    min_inflammations = data.min(axis=0) 
    std_inflammations = data.std(axis=0) 
    
    plt.figure(figsize=(12.0, 12.0))
    
    plt.subplot(2, 2, 1)
    plt.plot(ave_inflammations)
    plt.ylabel('average inflammations per day')
    plt.xlabel('days')
    
    plt.subplot(2, 2, 2)
    plt.plot(max_inflammations)
    plt.ylabel('max inflammations per day')
    plt.xlabel('days')    
    
    plt.subplot(2, 2, 3)
    plt.plot(min_inflammations)
    plt.ylabel('min inflammations per day')
    plt.xlabel('days')    
    
    plt.subplot(2, 2, 4)
    plt.plot(std_inflammations)
    plt.ylabel('standard deviation inflammations per day')
    plt.xlabel('days')    
```

```python
plot_properties(data) 
```
